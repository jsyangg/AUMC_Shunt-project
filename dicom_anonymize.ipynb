```python
import os
import string
import random
from tqdm import tqdm
import pandas as pd
```

```python
os.getcwd()
```

```python
head_savelist = []

for file in tqdm(os.listdir(r'D:/jsyang/aumc_medicalAI/BTS_head/')):
    
    folder_len = len(os.listdir(f'D:\jsyang/aumc_medicalAI/BTS_head/{file}/'))
    cnt = 0
    
    letters_set = string.ascii_letters + string.digits
    random_list = random.sample(letters_set,10)
    result = ''.join(random_list)
    
    head_savelist.append([file, result])
    
    for f in os.listdir(f'D:\jsyang/aumc_medicalAI/BTS_head/{file}/'):
        os.rename(f'D:\jsyang/aumc_medicalAI/BTS_head/{file}/{f}', f'D:\jsyang/aumc_medicalAI/BTS_head/{file}/{result}000{cnt}.dcm')
        cnt += 1
        
    os.rename(f'D:\jsyang/aumc_medicalAI/BTS_head/{file}', f'D:\jsyang/aumc_medicalAI/BTS_head/{result}')
```

```python
skull_savelist = []

for file in tqdm(os.listdir(r'D:/jsyang/aumc_medicalAI/BTS_skull/')):
    
    folder_len = len(os.listdir(f'D:\jsyang/aumc_medicalAI/BTS_skull/{file}/'))
    cnt = 0
    
    letters_set = string.ascii_letters + string.digits
    random_list = random.sample(letters_set,10)
    result = ''.join(random_list)
    
    skull_savelist.append([file, result])
    
    for f in os.listdir(f'D:\jsyang/aumc_medicalAI/BTS_skull/{file}/'):
        os.rename(f'D:\jsyang/aumc_medicalAI/BTS_skull/{file}/{f}', f'D:\jsyang/aumc_medicalAI/BTS_skull/{file}/{result}000{cnt}.dcm')
        cnt += 1
        
    os.rename(f'D:\jsyang/aumc_medicalAI/BTS_skull/{file}', f'D:\jsyang/aumc_medicalAI/BTS_skull/{result}')
```
